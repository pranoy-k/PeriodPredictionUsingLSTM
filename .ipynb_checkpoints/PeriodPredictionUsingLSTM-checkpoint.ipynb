### Import Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Loading the Data

```python
file = "data.txt"
X = []
Y = []
with open(file) as f:
    for line in f:
        temp = line.strip().split()
        Y.append(temp[0])
        X.append(temp[1:])
```

Formatting and Converting the data to Numpy Arrays

```python
## Checking the data format
X[0][1:5], Y[1:5]
```

```python
def convToNum(s):
    a,b = s.split('e')
    return float(a)*(10**float(b))

Y = np.array(list(map(convToNum, Y)))
temp = np.zeros((len(X),len(X[0])))
i = 0
for x in X:
    temp[i] = np.array(list(map(convToNum, x)))
    i += 1
X = temp
```

Showing one of the curve

```python
plt.plot(X[1][:100])
```

Randomly shuffling our data and verifying one of the datapoint

```python
np.random.seed(10)
np.random.shuffle(Y)
```
